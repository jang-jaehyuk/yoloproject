```python
def oxford2yolo(oxford_path, output_path, class_list = None, fierce = False):
    # input, output path 정의
    ANNO_path = os.path.join(oxford_path, 'annotations/xmls')
    IMAGE_path = os.path.join(oxford_path, 'images')
    OUTPUT_IMAGE_path = os.path.join(output_path, 'images')
    OUTPUT_ANNO_path = os.path.join(output_path, 'labels')

    if not os.path.exists(output_path):
        os.mkdir(output_path)
        os.mkdir(OUTPUT_IMAGE_path)
        os.mkdir(OUTPUT_ANNO_path)

    idx = len(os.listdir(OUTPUT_IMAGE_path)) + 1
    remove_cnt = 0

    dog_list = os.listdir(ANNO_path)
    for f in dog_list:
        file_path = os.path.join(ANNO_path, f)

        # ===== xml파일 parsing ===== # 
        tree = ET.parse(file_path)
        root = tree.getroot()

        file_name = root.find('filename').text
        class_name = file_name.split('_')[0]

        if class_name not in class_list:
            remove_cnt += 1
            continue
        
        class_id = class_list.index(class_name)

        if fierce:
            class_name = 'fierce' if class_name in [
                'Staffordshire Bull Terrier',
                'American Pit Bull Terrier'
            ] else 'normal'
            
            class_id = 0 if class_name == 'normal' else 'fierce'
        
        size = root.find('size')

        if size is None:
            remove_cnt += 1
            continue

        img_width = int(size.find('width').text)
        img_height = int(size.find('height').text)

        with open(os.path.join(OUTPUT_ANNO_path, f'{str(idx)}.txt'), 'w') as f:
            for obj in root.findall('object'):
                # bounding box
                box = obj.find('bndbox')
                xmin = int(box.find('xmin').text)
                ymin = int(box.find('ymin').text)
                xmax = int(box.find('xmax').text)
                ymax = int(box.find('ymax').text)

                # annotation 변환
                x, y, width, height = coord2YOLOcoord(img_width, img_height, xmin, ymin, xmax, ymax)
                text = '{} {} {} {} {}'.format(class_id, x, y, width, height)

                f.write(text + '\n')

        shutil.copy(os.path.join(IMAGE_path, file_name), os.path.join(OUTPUT_IMAGE_path, f'{str(idx)}.jpg'))

        idx += 1
    
    print('OXFORD: {}개의 이미지가 class_list에 존재하지 않아 제거되었습니다.'.format(remove_cnt))
    print('OXFORD: {}개의 이미지가 저장되었습니다.'.format(idx - 1))
        



def stanford2yolo(stanford_path, output_path, class_list = None, fierce = False):
    # input, output path 정의
    ANNO_path = os.path.join(stanford_path, 'Annotation')
    IMAGE_path = os.path.join(stanford_path, 'Images')
    OUTPUT_IMAGE_path = os.path.join(output_path, 'images')
    OUTPUT_ANNO_path = os.path.join(output_path, 'labels')

    if not os.path.exists(output_path) or not os.path.exists(OUTPUT_IMAGE_path):
        os.mkdir(output_path)
        os.mkdir(OUTPUT_IMAGE_path)
        os.mkdir(OUTPUT_ANNO_path)

    idx = len(os.listdir(OUTPUT_IMAGE_path)) + 1
    remove_cnt = 0

    # 견종별 path list
    dog_path_list = os.listdir(ANNO_path)
    for dog_path in dog_path_list:
        # 개별 종 path list
        species_path = os.path.join(ANNO_path, dog_path)
        dog_list = os.listdir(species_path)

        for dog in dog_list:
            # 개별 개체 path list
            file_path = os.path.join(species_path, dog)

            # ===== xml파일 parsing ===== # 
            tree = ET.parse(file_path)
            root = tree.getroot()
            
            # image size
            img_size = root.find('size')
            img_width = int(img_size.find('width').text)
            img_height = int(img_size.find('height').text)

            remove = []
            for obj in root.findall('object'):
                name = obj.find('name').text
                r = name in class_list
                remove.append(r)
            
            save = all(remove)
            if not save:
                remove_cnt += 1
                continue

            # object
            with open(os.path.join(OUTPUT_ANNO_path, f'{str(idx)}.txt'), 'w') as f:
                for obj in root.findall('object'):
                    class_name = obj.find('name').text

                    class_id = class_list.index(class_name)

                    # fierce == True이면 맹견 or not의 이진분류
                    if fierce:
                        class_name = 'fierce' if class_name in [
                            'Staffordshire_bullterrier',
                            'Rottweiler',
                            'American_Staffordshire_terrier'
                        ] else 'normal'

                        class_id = 0 if class_name == 'normal' else 1
                    
                    # bounding box
                    box = obj.find('bndbox')
                    xmin = int(box.find('xmin').text)
                    ymin = int(box.find('ymin').text)
                    xmax = int(box.find('xmax').text)
                    ymax = int(box.find('ymax').text)

                    # annotation 변환
                    x, y, width, height = coord2YOLOcoord(img_width, img_height, xmin, ymin, xmax, ymax)
                    text = '{} {} {} {} {}'.format(class_id, x, y, width, height) 

                    # txt file로 저장
                    f.write(text + '\n')
        
            # image 복사
            shutil.copy(file_path.replace('Annotation', 'Images') + '.jpg', os.path.join(OUTPUT_IMAGE_path, f'{str(idx)}.jpg'))
            
            idx += 1

    print('STANFORD: {}개의 이미지가 class_list에 존재하지 않아 제거되었습니다.'.format(remove_cnt))
    print('STANFORD: {}개의 이미지가 저장되었습니다.'.format(idx - 1))


    
def train_valid_split(path, valid_size = 0.1, image = True, indices = None):
    train_path = os.path.join(path, 'train')
    valid_path = os.path.join(path, 'valid')
    
    if not indices:
        file_list = os.listdir(path) 
        file_range = [f.split('.')[0] for f in file_list]
        valid_cnt = int(len(file_list) * valid_size)
        random_idx = np.random.permutation(file_range)
        valid_indices = random_idx[:valid_cnt]
        train_indices = random_idx[valid_cnt:]
    
    else: 
        train_indices, valid_indices = indices

    if not os.path.exists(train_path):
        os.mkdir(train_path)
    if not os.path.exists(valid_path):
        os.mkdir(valid_path)

    # validation data move
    for idx in valid_indices:
        if image:
            src_path = os.path.join(path, '{}.jpg'.format(idx))
            tgt_path = os.path.join(valid_path, '{}.jpg'.format(idx))
        else:
            src_path = os.path.join(path, '{}.txt'.format(idx))
            tgt_path = os.path.join(valid_path, '{}.txt'.format(idx))

        shutil.move(src_path, tgt_path)
    
    # train data move
    for idx in train_indices:
        if image:
            train_src_path = os.path.join(path, '{}.jpg'.format(idx))
            train_tgt_path = os.path.join(train_path, '{}.jpg'.format(idx))
        else:
            train_src_path = os.path.join(path, '{}.txt'.format(idx))
            train_tgt_path = os.path.join(train_path, '{}.txt'.format(idx))

        shutil.move(train_src_path, train_tgt_path)
    
    return (train_indices, valid_indices)

def coord2YOLOcoord(img_width, img_height, xmin, ymin, xmax, ymax):
    x_center = round(((xmin + xmax) / 2)/img_width, 7)
    y_center = round(((ymin + ymax) / 2)/img_height, 7)
    width = round((xmax - xmin)/img_width, 7)
    height = round((ymax - ymin)/img_height, 7)

    return x_center, y_center, width, height
```

```python
import shutil
import wget
import argparse
import os
import xml.etree.ElementTree as ET
import numpy as np

def make_custom_dataset(config):
    STANFORD_DOGS_DATA_URL = {
        'images' : 'http://vision.stanford.edu/aditya86/ImageNetDogs/images.tar',
        'annotations' : 'http://vision.stanford.edu/aditya86/ImageNetDogs/annotation.tar'
    }
    TSINGHUA_DATA_URL = {
        'images' : 'https://cloud.tsinghua.edu.cn/f/80013ef29c5f42728fc8/?dl=1',
        'annotations' : 'https://cg.cs.tsinghua.edu.cn/ThuDogs/low-annotations.zip'
    }
    OXFORD_PET_DATA_RUL = {
        'images' : 'https://thor.robots.ox.ac.uk/~vgg/data/pets/images.tar.gz',
        'annotations' : 'https://thor.robots.ox.ac.uk/~vgg/data/pets/annotations.tar.gz'
    }

    # root directory로 이동
    os.chdir(config.root_dir)
    dir_list = os.listdir(config.root_dir)

    # raw_dataset directory가 없다면 생성
    if 'raw_datasest' not in dir_list:
        os.mkdir('raw_dataset')

    # raw_dataset directory로 이동
    os.chdir('raw_dataset')

    # data download(standford)
    os.mkdir('./stanford')
    os.chdir('./stanford')
    image_file_name = wget.download(STANFORD_DOGS_DATA_URL['images'])
    anno_file_name = wget.download(STANFORD_DOGS_DATA_URL['annotations'])
    shutil.unpack_archive(image_file_name)
    shutil.unpack_archive(anno_file_name)

    # data download(tsinghua)
    # os.chdir('..')
    # os.mkdir('./tsinghua')
    # os.chdir('./tsinghua')
    # image_file_name = wget.download(TSINGHUA_DATA_URL['images'])
    # anno_file_name = wget.download(TSINGHUA_DATA_URL['annotations'])
    # shutil.unpack_archive(image_file_name)
    # shutil.unpack_archive(anno_file_name)

    # data download(oxford)
    os.chdir('..')
    os.mkdir('./oxford')
    os.chdir('./oxford')
    image_file_name = wget.download(OXFORD_PET_DATA_RUL['images'])
    anno_file_name = wget.download(OXFORD_PET_DATA_RUL['annotations'])
    shutil.unpack_archive(image_file_name)
    shutil.unpack_archive(anno_file_name)

    os.chdir(config.root_dir)
    
    # stanford data processing
    stanford_class_list = os.listdir('./raw_dataset/stanford/Images')
    stanford_class_list = [dog.split('-')[-1] for dog in stanford_class_list]
    stanford2yolo('./raw_dataset/stanford', './custom_dataset', class_list = stanford_class_list, fierce = config.fierce)
    # oxford data processing
    oxford_class_list = os.listdir('./raw_dataset/oxford/images')
    oxford_class_list = [dog.split('_')[0] for dog in oxford_class_list]
    oxford_class_list = list(set(oxford_class_list))
    oxford2yolo('./raw_dataset/oxford', './custom_dataset', class_list = oxford_class_list, fierce = config.fierce)

    # image split
    indices = train_valid_split('./custom_dataset/images')
    # annotation split
    train_valid_split('./custom_dataset/labels', image = False, indices = indices)

```
